Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordenadas do launch site
launch_site_lat = 28.562302  # Substituir pelo launch site que você está usando
launch_site_lon = -80.577356

# Coordenadas do ponto mais próximo da costa (obtido usando MousePosition)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calcular a distância até a costa
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

print(f'Distância até a costa: {distance_coastline:.2f} km')
```

```python
from folium.features import DivIcon

# Coordenadas do launch site
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordenadas do ponto mais próximo da costa (exemplo obtido com MousePosition)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calcular a distância entre o launch site e a costa
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Coordenada para o marcador da costa
coastline_coordinate = [coastline_lat, coastline_lon]

# Criar o marcador da costa com a distância mostrada no ícone
distance_marker = folium.Marker(
    location=coastline_coordinate,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 14px; color:#d35400;"><b>{distance_coastline:.2f} KM</b></div>',
    )
)

# Adicionar o marcador da costa no mapa
distance_marker.add_to(site_map)

# Opcional: adicionar uma linha ligando o launch site à costa
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], coastline_coordinate],
    color='blue',
    weight=2,
    opacity=0.6
).add_to(site_map)

# Exibir o mapa
site_map
```
